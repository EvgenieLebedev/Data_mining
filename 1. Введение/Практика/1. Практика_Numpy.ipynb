10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
matrix = np.random.randint(0, 10, size=(10, 10)) #случайные целые числа от 0 до 9.
#matrix = np.random.rand(10, 10) # альтернативный путь
row_means = np.mean(matrix, axis=1, keepdims=True)


print(f'среднее по строкам= {row_means.transpose()}' , f'\n матрица_исходная =\n {matrix}', f'\n матрица - среднее =\n {matrix - row_means}')
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
matrix = np.random.rand(5, 5) 

max_values = np.max(matrix, axis=1, keepdims=True)
matrix_with_max = np.hstack((matrix, max_values))

min_values = np.min(matrix, axis=1, keepdims=True)
matrix_with_max_and_min = np.hstack((matrix_with_max, min_values))
 
print(f'матрица исходная = \n {matrix},\n максимальные строки  =  {max_values.transpose()},\n минимальные строки  =  {min_values.transpose()}, \n матрица итоговая =  \n {matrix_with_max_and_min}')
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу чётого произведения.  

```python
vector = np.arange(1, 52)

odd_numbers = vector[vector % 2 == 1]
even_numbers = vector[vector % 2 == 0]

odd_matrix = odd_numbers.reshape(-1, 1)
even_matrix = even_numbers.reshape(-1, 1)

result_matrix = odd_matrix.dot(even_matrix.T)

try:
    inverse_matrix = np.linalg.inv(result_matrix)
except np.linalg.LinAlgError:
    inverse_matrix = None

print("Матрица нечетных чисел:")
print(odd_matrix.transpose())

print("\nМатрица четных чисел:")
print(even_matrix.transpose())

print("\nПроизведение матриц:")
print(result_matrix)

print("\nОбратная матрица произведения (если существует):")
print(inverse_matrix) #None - матрица не существует
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
matrix = np.zeros((8, 8), dtype=int)
matrix[1::2, ::2] = 1  # четные строки нечетными столбцами
matrix[::2, 1::2] = 1  # нечетные строки четными столбцами

matrix
```
